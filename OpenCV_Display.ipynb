```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

img = cv2.imread('/image.jpg') ### add your image here
plt.imshow(img)
plt.show()
```

Image loads nicely but... doesn't look as expected' 
OpenCV will load the images as BGR not RGB so we need to convert them


```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

That is better! Now lets convert to HSV

```python
img_hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
plt.imshow(img_hsv)
plt.show()
```

Now lets plot the images alongside each other - we will use subplots

```python
img_hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

plt.figure(figsize=(10, 3))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)

plt.subplot(1, 2, 2)
plt.imshow(img_hsv)

plt.show()

```

OpenCV HSV range is: H: 0 to 179 S: 0 to 255 V: 0 to 255

```python
## https://giusedroid.blogspot.co.uk/2015/04/using-python-and-k-means-in-hsv-color.html

hue, sat, val = img_hsv[:,:,0], img_hsv[:,:,1], img_hsv[:,:,2]

plt.figure(figsize=(10,8))
plt.subplot(311)                             #plot in the first cell
plt.subplots_adjust(hspace=.5)
plt.title("Hue")
plt.hist(np.ndarray.flatten(hue), bins=180)
plt.subplot(312)                             #plot in the second cell
plt.title("Saturation")
plt.hist(np.ndarray.flatten(sat), bins=128)
plt.subplot(313)                             #plot in the third cell
plt.title("Luminosity Value")
plt.hist(np.ndarray.flatten(val), bins=128)
plt.show()
```
